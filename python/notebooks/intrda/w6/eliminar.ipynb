### Pregunta 3
Hacemos una columna con la suma de los productos de cada boleta

Primero, creamos una columna que corresponde a la suma de los precios de los productos que componen cada boleta

```python
pro = ['Producto 1','Producto 2','Producto 3','Producto 4','Producto 5',
             'Producto 6','Producto 7','Producto 8','Producto 9','Producto 10']
suma_bol = []
for i in range(1, len(boletas)+1):
    suma = 0
    for producto in pro:
        if pd.isna(boletas[producto][i]) != True:
            suma += productos.loc[productos['Producto'] == boletas[producto][i]].reset_index()['Precio'][0]
        else:
            pass
    suma_bol.append(suma)

boletas['Suma_productos'] = suma_bol
boletas
```

ahora creamos una columna con la diferencia entre la columna valor boleta y la columna con el precio de la boleta calculado por nosotros

```python
boletas['Diferencia'] = boletas['Valor Boleta'] - boletas['Suma_productos']
boletas
```

```python
#y contamos la cantidad de registros en esta nueva columna 'Diferencia' que sean mayores que 0.
boletas[boletas['Diferencia'] > 0]['Diferencia'].count()
```

### Pregunta 4

Buscar el producto que tenga el valor de la diferencia y lo colocamos en la columna 'Producto 1'.

```python
#Como hay diferencias, se despliega Dataframe original para filas con diferencia != 0
boletas.loc[boletas['Diferencia'] != 0]
```

```python
#Realizaremos primero el ejercicio solamente para un registro, el de la fila 98. Queremos obtener el nombre del producto
#que tiene el mismo precio que el valor guardado en 'Diferencia'.
precio = boletas['Diferencia'][98]
producto = productos.loc[productos['Precio'] == precio].reset_index()['Producto'][0]
producto
```

```python
#Ahora lo hacemos para todos los registros faltantes en 'Producto 1'.
#Buscamos cada producto faltante en 'Producto 1' y lo reemplazamos en la columna 'Producto 1'
producto1 = []
for i in range(1, len(boletas)+1):
    if boletas['Diferencia'][i] != 0:
        precio = boletas['Diferencia'][i]
        producto = productos.loc[productos['Precio'] == precio].reset_index()['Producto'][0]
        producto1.append(producto)
    else:
        producto1.append(boletas['Producto 1'][i])

boletas['Producto 1'] = producto1
```

```python
print("¿Existen elementos vacios en producto 1? ", boletas['Producto 1'].isnull().values.any())
```
