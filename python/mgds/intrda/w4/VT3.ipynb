```python
import pandas as pd
import numpy as np
import re
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('VT3.xlsx', sheet_name="Respuestas de formulario 1")
df.head(5)
```

```python
df = df.replace("Excelente o muy por sobre lo esperado", 5)
df = df.replace("Muy bueno", 4)
df = df.replace("Satisfactorio o dentro de lo esperado", 3)
df = df.replace("Regular", 2)
df = df.replace("Insatisfactorio o por debajo de lo esperado", 1)
df = df.replace("Este soy yo", np.nan)
```

```python
df = df[df["Elige tu equipo"]=="Equipo 9"]
df
```

```python
df = df.dropna(axis=1,how='all')
df
```

```python
df = df.drop_duplicates(subset="Correo UC", keep="last")
df
```

```python
estudiantes_grupo = []
for c in df.columns[4:]:
    estudiante_grupo = c.split(" [")[1][:-1].replace("].","")
    if estudiantes_grupo not in estudiantes_grupo:
        estudiantes_grupo.append(estudiante_grupo)
estudiantes_grupo =sorted(list(set(estudiantes_grupo)))
estudiantes_grupo
```

```python
estudiante_grupo = estudiantes_grupo[1]
print("Estudiante del grupo:", estudiante_grupo)
```

```python
relevant_values = []
for c in df.columns:
    if estudiante_grupo in c:
        relevant_values.append(c)
relevant_values
```

```python
for c in df.columns[4:]:
    df[c] = df[df[c].notna()][c].astype(float)
```

```python
notas_dict = {c:df[c].mean() for c in relevant_values}
```

```python
for k in notas_dict.keys():
    print(k, notas_dict[k])
```
