```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset("tips")
```

```python
len(tips)
```

```python
tips.head(5)
```

```python
plt.scatter(tips.total_bill, tips.tip)
```

```python
plt.scatter(tips.total_bill, tips.tip, color="red", marker='s', alpha=0.3)
plt.title("Cuenta Vs Propina")
plt.xlabel("Cuenta")
plt.ylabel("Propina")
```

```python
ax = sns.scatterplot(x='total_bill', y="tip", data=tips)
```

```python
ax = sns.scatterplot(x="total_bill", y="tip", data=tips, color='red', marker='s', alpha=0.3)
```

```python
smoker_num = tips.smoker.astype("category").cat.codes
time_num = tips.time.astype("category").cat.codes
```

```python
sns.set_style("whitegrid")
ax = sns.scatterplot(x="total_bill", y="tip", hue=time_num, 
                     size="size", sizes=(20,250), hue_norm=(0,1.5),
                     alpha=0.7,
                     legend="full", data=tips)
```
