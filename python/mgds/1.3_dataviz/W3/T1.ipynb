```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset("tips")
df.head(5)
```

```python
df.tail(5)
```

```python
df.describe()
```

```python
tip_por_dia = df.groupby('day')['tip'].mean()
```

```python
print(tip_por_dia)
```

```python
ax = tip_por_dia.plot.bar(color='red', title='Propina promedio por dia')
ax.set_xlabel('Dia de la semana')
ax.set_xlabel('Propina promedio')
```

```python
MH_por_dia = df.groupby(['day', 'sex'])['tip'].count()
```

```python
MH_por_dia
```

```python
ax = MH_por_dia.plot.bar(stacked=True, title='Cantidad de mesas por dia y sexo')
ax.set_xlabel('Dia de la semana')
ax.set_xlabel('Cantidad de mesas')
```

```python
wide_MH_por_dia = MH_por_dia.unstack(level=1)
wide_MH_por_dia
```

```python
ax = wide_MH_por_dia.plot.bar(stacked=True, title='Cantidad de clientes por dia y sexo')
ax.set_xlabel('Dia de la semana')
ax.set_ylabel('Cantidad de clientes')
```

```python
fmri = sns.load_dataset("fmri")
len(fmri)
```

```python
fmri.head(5)
```

```python
subject13 = fmri.loc[(fmri.subject=='s13') & (fmri.event == 'cue') & (fmri.region == 'frontal')]
print(subject13)
```

```python
subject13 = subject13.sort_values(by='timepoint')
subject13
```

```python
sns.set()
sns.set_style('whitegrid')

plt.plot(subject13.timepoint, subject13.signal, marker='o', linestyle='')
```

```python
ax = sns.lineplot(x='timepoint', y='signal', data=subject13, linewidth=0, marker='o')
```

```python
ax = sns.lineplot(x='timepoint', y='signal', color='green', data=fmri, linewidth=0, marker='s',err_style='bars')
```
