```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset("tips")
```

```python
len(tips)
```

```python
tips.head(5)
```

```python
#Histograma
ax = sns.displot(tips.tip)
```

```python
sns.displot(tips.tip, kde=True, color="red", bins=25)
```

```python
sns.histplot(tips, x="tip", kde = False, hue="time")
```

```python
f, axes = plt.subplots(1,2,figsize=(12,4), sharex=True)
sns.histplot(tips[tips.time=="Lunch"], x="tip", kde=True, color="red", label="Propinas en Almuerzo", ax=axes[0])
sns.histplot(tips[tips.time=="Dinner"], x="tip", kde=True, color="skyblue", label="Propinas en Cena", ax=axes[1])
```

```python
sns.displot(tips.tip, kind="kde", color="red")
```

```python
sns.displot(tips, x="tip", kind="kde", hue="time")
```

```python
ax = sns.boxplot(tips.tip)
```

```python
ax = sns.boxplot(tips.tip, orient = "v", color="red")
```

```python
sns.boxplot(x="day", y="tip", data=tips, palette="Set3")
```

```python
ax = sns.violinplot(x=tips["total_bill"])
```

```python
ax = sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
ax = sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips, palette="muted")
```
