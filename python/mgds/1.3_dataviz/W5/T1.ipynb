```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataset_vuelos = pd.read_csv('../data/flights.csv').pivot(index='year', columns='month', values='passengers')
dataset_vuelos.head(5)
```

```python
plt.rcParams['figure.figsize'] = (15.0, 8.0)
ax = sns.heatmap(dataset_vuelos)
```

```python
ax = sns.heatmap(dataset_vuelos, cmap='coolwarm', linewidths=.5, square=True)
```

```python
ax = sns.heatmap(dataset_vuelos, annot=True, fmt="d", annot_kws={'size':10}, cmap='coolwarm', linewidths=.5)
```

```python
penguins = sns.load_dataset("penguins")
penguins.head(5)
```

```python
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species"
)
```

```python
sns.kdeplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    levels=3,
    thresh=.2
)
```

```python
sns.kdeplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    levels=6,
    thresh=.2,
    fill=True
)
```

```python
ax = penguins.plot.hexbin(
    x="bill_length_mm",
    y='bill_depth_mm',
    reduce_C_function=np.sum,
    gridsize=16,
    cmap="viridis"
    )
```
